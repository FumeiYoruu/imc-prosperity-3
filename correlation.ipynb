```python
def get_correlation(x: np.ndarray, y: np.ndarray) -> float:
    assert x.shape == y.shape

    correlation_matrix = np.corrcoef(x, y)

    return correlation_matrix[0, 1]
```

```python
def compare_datas(paths_a, paths_b):
    correlations = []

    for path_from_a in paths_a:
        for path_from_b in paths_b:
            df_from_a = pd.read_csv(path_from_a, sep = ';')
            df_from_b = pd.read_csv(path_from_b, sep = ';')

            for product_from_a in df_from_a['product'].unique():
                for product_from_b in df_from_b['product'].unique():
                    data_from_a = df_from_a.loc[df_from_a['product'] == product_from_a]['mid_price'].to_numpy()
                    data_from_b = df_from_b.loc[df_from_b['product'] == product_from_b]['mid_price'].to_numpy()

                    if data_from_a.shape == data_from_b.shape:
                        correlations.append(
                            (
                                get_correlation(data_from_a, data_from_b),
                                (path_from_a, product_from_a),
                                (path_from_b, product_from_b)
                            )
                        )

    correlations = sorted(correlations, reverse = True)

    return correlations
```

```python
correlations_3_2 = compare_datas(paths3, paths2)
correlations_3_1 = compare_datas(paths3, paths1)
```

```python
for data in correlations_3_2:
    corr, from3, from2 = data

    if from3[1] != 'VOLCANIC_ROCK':
        continue

    if corr > 0.8:
        df_from3 = pd.read_csv(from3[0], sep = ';')
        df_from2 = pd.read_csv(from2[0], sep = ';')

        data_from3 = df_from3.loc[df_from3['product'] == from3[1]]['mid_price'].to_numpy()
        data_from2 = df_from2.loc[df_from2['product'] == from2[1]]['mid_price'].to_numpy()

        data_from3 = (data_from3 - data_from3.min()) / (data_from3.max() - data_from3.min())
        data_from2 = (data_from2 - data_from2.min()) / (data_from2.max() - data_from2.min())

        plt.plot(data_from3)
        plt.plot(data_from2)

        print(corr)
        print(from3)
        print(from2)

        plt.show()
```
